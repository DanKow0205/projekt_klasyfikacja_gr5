```python
df.drop("Over18", axis=1, inplace=True)
```

```python
df.MaritalStatus.value_counts()
```

```python
df = df.join(pd.get_dummies(df["MaritalStatus"], prefix = "MaritalStatus"))
df = df.drop("MaritalStatus_Single", axis=1)
```

```python
df.PerformanceRating.value_counts()
```

```python
df = df.join(pd.get_dummies(df["PerformanceRating"], prefix = "PerformanceRating")).drop("PerformanceRating", axis=1)
```

```python
df.OverTime.value_counts()
```

```python
df["OverTime"] = [1 if x == "Yes" else 0 for x in df["OverTime"]]
```

```python
import plotly.express as px
for x in ["MonthlyIncome", "MonthlyRate", "NumCompaniesWorked", "PercentSalaryHike"]:
  fig = px.histogram(df, x=x)
  fig.show()
```
